# **Text-Encoding**

# Text Encoding in NLP

## Why Encoding Matters
Computers store only numbers, so text characters must be converted into numeric form using encoding before processing.

## Encoding Standards
- ASCII  
  Supports only English characters and is now outdated.

- Unicode  
  A universal standard supporting most world languages.
  - UTF-8: Uses variable length encoding and is memory efficient for English text.
  - UTF-16: Uses fixed length encoding and is better for many non-English characters.

## UTF-8 vs UTF-16
- English and ASCII characters: UTF-8 uses less memory
- Non-ASCII characters: UTF-16 may use memory more efficiently

## Python and Encoding
- Default encoding in Python is UTF-8
- Text operations:
  - Encode: convert string to bytes
  - Decode: convert bytes back to string

## Key Takeaway
Understanding text encoding is essential for multilingual NLP tasks.  
Always check the encoding format before processing text.


````python
# create a string
amount = u"50"
print('Default string: ', amount, '\n', 'Type of string', type(amount), '\n')

# encode to UTF-8 byte format
amount_encoded = amount.encode('utf-8')
print('Encoded to UTF-8: ', amount_encoded, '\n', 'Type of string', type(amount_encoded), '\n')


# sometime later in another computer...
# decode from UTF-8 byte format
amount_decoded = amount_encoded.decode('utf-8')
print('Decoded from UTF-8: ', amount_decoded, '\n', 'Type of string', type(amount_decoded), '\n')
````

# Regular Expressions (Regex)  Introduction

## What is a Regular Expression?
A **regular expression (regex)** is a sequence of characters that defines a **pattern**.  
It is used to **find, extract, substitute, or manipulate** substrings in a text.

## Why Regex is Important
- Extract **hashtags** from tweets (`#mumbai`, `#bangalore`, etc.)
- Extract **phone numbers** or other patterns from text
- Perform **string replacements** and other text manipulations

## Key Points
- Regex allows you to **identify patterns** in text efficiently.
- Regex has its own syntax and compiler.
- Supported by almost all popular programming languages, including Python.

## Regex in Python
- Use the `re` module for regular expressions.
- `re.search(pattern, string)`:
  - Returns a **RegexObject** if the pattern is found.
  - Returns **None** if the pattern is not found.
- `'match string'` and `'extract string'` are used interchangeably in practice problems.

# Question 1: Check Presence of 'education'

**Description:**  
Check whether the word `'education'` is present in the sentence:  
*"The roots of education are bitter, but the fruit is sweet."*  
Use `re.search()` to return `True` if present, else `False`.


````python
# Question 1: Python Code
import re

string = 'The roots of education are bitter, but the fruit is sweet.'

# regex pattern to check if 'education' is present
pattern = 'education'

# check whether pattern is present in string or not
result = re.search(pattern, string)

if result != None:
    print(True)
else:
    print(False)
````

# Regex Exercise  Match 'tree' or 'trees'

**Description:**  
Write a regular expression that matches the word `'tree'` or `'trees'` in a given text.

**Sample positive cases:**
- 'The tree stands tall.'
- 'There are a lot of trees in the forest.'

**Sample negative cases:**
- 'The boy is heading for the school.'

# Regex Exercise  Match 'x', 'xy', 'xz', 'xyz'

**Description:**  
Write a regular expression that matches the following words:  
- `x`  
- `xy`  
- `xz`  
- `xyz`  

Do **not** match:  
- `Xyyz`, `Xyzz`, `Xyy`, `Xzz`, `Yz`


````python
# Python Code
import re

# input string (replace with test string)
string = 'xyz'  # Example: change to 'x', 'xy', 'xz', etc. for testing

# regex pattern to match 'x', 'xy', 'xz', 'xyz' only
pattern = r'x(y)?(z)?$'

# check whether pattern is present in string
result = re.search(pattern, string)

# evaluate result
if result != None:
    print(True)
else:
    print(False)
````

# Regular Expressions  Quantifiers II

## Recap of Previous Quantifiers
- `?`  Matches the preceding character **0 or 1 time** (optional)  
- `*`  Matches the preceding character **0 or more times**  
- `+`  Matches the preceding character **1 or more times** (at least once)

**Difference between `*` and `+`:**  
- `*`  character may be absent  
- `+`  character must appear at least once

## Fixed-Occurrence Quantifiers
Sometimes you need to match a character **exactly** a number of times or within a range:

1. `{m, n}`  Matches the preceding character **m to n times**  
2. `{m,}`  Matches the preceding character **m to infinite times**  
3. `{, n}`  Matches the preceding character **0 to n times**  
4. `{n}`  Matches the preceding character **exactly n times**

**Notes:**  
- Avoid spaces inside `{m,n}`  write `{m,n}` not `{m, n}`  
- These can **replace other quantifiers**:
  - `?`  `{0,1}`  
  - `*`  `{0,}`  
  - `+`  `{1,}`

# Regex Comprehension Exercises

## 1. Whitespace in Regex
- Whitespace includes spaces, tabs (`\t`) or newlines (`\n`)  
- Example: `"James Allen"` matches exactly that name including the space  
- `" +"` matches **one or more spaces**


## 2. Grouping with Parentheses
- `(abc){1,3}`  Matches `abc`, `abcabc`, `abcabcabc`  
- `(010)+`  Matches `010`, `010010`, `010010010`,   
- Parentheses allow quantifiers to repeat **groups of characters**


````python
import re
import sys

string = sys.stdin.read()

# Match '23' one or more times followed by '78' one or more times
pattern = r'(23)+?(78)+'

result = re.search(pattern, string)
print(result != None)
````

## 3. Pipe Operator `|` (OR)
- `(d|g)one`  matches `done` or `gone`  
- `(ICICI|HDFC) Bank`  matches `ICICI Bank` or `HDFC Bank`  
- Multiple pipes: `(0|1|2){2}`  matches any two-digit string composed of 0,1,2


````python
import re
import sys

string = sys.stdin.read()

# Match Basketball, Baseball, Volleyball, Softball, Football
pattern = r'(Basketball|Baseball|Volleyball|Softball|Football)'

result = re.search(pattern, string)
print(result != None)
````

## 4. Escaping Special Characters
- `?`, `*`, `+`, `(`, `)`, `{` are special characters in regex  
- Escape them with `\` to match **literally**  
- Example:  
  - `\?`  matches literal question mark  
  - `\+`  matches literal plus  
  - `\(78\)`  matches literal `(78)`


````python
import re
import sys

string = sys.stdin.read()

# Match multiplication equations containing '*'
pattern = r'\*'

result = re.search(pattern, string)
print(result != None)
````

## 5. Regex Flags
- `re.I`  Ignore case  
- `re.M`  Multi-line search  
- Multiple flags: `flags = re.I | re.M`  

Example:
